```python
q3 = weekly_menu['VST_RAT'].quantile(0.75) # df['국어'].quantile(0.75) 처럼 특정 열만 적용 가능
q1 = weekly_menu['VST_RAT'].quantile(0.25)
iqr = q3 - q1
q1, q3, iqr
```

```python
def outlier(weekly_menu):
    if (weekly_menu['VST_RAT'] > (q3 + 1.5 * iqr)) or (weekly_menu['VST_RAT'] < (q1 - 1.5 * iqr)):
        return True
    else:
        return False
import numpy as np
weekly_menu['이상치여부'] = np.where(((weekly_menu['VST_RAT'] > (q3 + 1.5 * iqr)) | (weekly_menu['VST_RAT'] < (q1 - 1.5 * iqr))), True, False)
```

```python
weekly_menu['이상치여부'] = weekly_menu['VST_RAT'] == 100.00
```

```python
weekly_menu['이상치여부'].value_counts()
```

```python
# 이상치 제거 후 확인
new_w_m = weekly_menu.loc[weekly_menu['이상치여부'] == False]
del new_w_m['이상치여부']
```

```python
# 필요한 데이터만 뽑아내기
new_w_m = new_w_m[['DAY_CD', 'INDTY_CGR_NM', 'RANK']]
```
